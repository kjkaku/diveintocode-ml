### 【問題6】図の作成
データは可視化して確認することが大切です。次の3つを行ってください。作図のライブラリはMatplotlibの他seabornもよく使われます。

- ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
- 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する
- 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する
箱ひげ図とバイオリン図は似た図を表示します。これらはどう違い、それぞれどのような利点があるのかについて説明してください。

```python
#ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
label = ["setosa", "virgiclor", "virginica"]
plt.title("Pie chart of Fisher's iris")
plt.pie(df.groupby("Species").size(), labels=label, autopct="%1.1f%%")
plt.show()
```

```python
# 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する

# sepal_lengthの箱ひげ図

sepal_length_data = (df[df["Species"] == 0]["sepal_length"],
                     df[df["Species"] == 1]["sepal_length"],
                     df[df["Species"] == 2]["sepal_length"])

sepal_width_data = (df[df["Species"] == 0]["sepal_width"],
                    df[df["Species"] == 1]["sepal_width"],
                    df[df["Species"] == 2]["sepal_width"])

petal_length_data = (df[df["Species"] == 0]["petal_length"],
                     df[df["Species"] == 1]["petal_length"],
                     df[df["Species"] == 2]["petal_length"])

petal_width_data = (df[df["Species"] == 0]["petal_width"],
                    df[df["Species"] == 1]["petal_width"],
                    df[df["Species"] == 2]["petal_width"])

fig, ax = plt.subplots()

bp = ax.boxplot(sepal_length_data)
ax.set_xticklabels(label)
plt.ylabel("sepal_length")
plt.grid()

fig, ax = plt.subplots()
bp = ax.boxplot(sepal_width_data)
ax.set_xticklabels(label)
plt.ylabel("sepal_width")
plt.grid()

fig, ax = plt.subplots()
bp = ax.boxplot(petal_length_data)
ax.set_xticklabels(label)
plt.ylabel("petal_length")
plt.grid()

fig, ax = plt.subplots()
bp = ax.boxplot(petal_width_data)
ax.set_xticklabels(label)
plt.ylabel("petal_width")
plt.grid()
```

```python
#特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する 箱ひげ図とバイオリン図は似た図を表示します。
#これらはどう違い、それぞれどのような利点があるのかについて説明してください。

fig, ax = plt.subplots()
plt.violinplot(sepal_length_data)
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(label)
plt.ylabel("sepal_length")
plt.show()

fig, ax = plt.subplots()
plt.violinplot(sepal_width_data)
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(label)
plt.ylabel("sepal_width")
plt.show()

fig, ax = plt.subplots()
plt.violinplot(petal_length_data)
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(label)
plt.ylabel("petal_length")
plt.show()

fig, ax = plt.subplots()
plt.violinplot(petal_width_data)
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(label)
plt.ylabel("petal_width")
plt.show()
```

#### 箱ひげ図とバイオリン図の違いについて
箱ひげ図には平均値、中央値や四分位範囲という要約統計量のみが含まれるのに対し、バイオリン図にはすべての分布データが含まれる。複数のピークがあるデータなどにはバイオリン図の方が表現しやすい場合がある。

```python
# 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species", size=3.5)
plt.show()
```

```python
# ４つの特徴量に関する相関係数行列を作る
corr_mat = df.corr(method='pearson')
corr_mat
```

```python
# 相関係数行列のヒートマップを作る
sns.heatmap(corr_mat,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に値を表示
            fmt='.1f',
            xticklabels=corr_mat.columns.values,
            yticklabels=corr_mat.columns.values
           )
plt.show()
```
