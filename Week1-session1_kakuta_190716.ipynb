　秀吉から褒美を下される際、何を希望するか尋ねられた新左衛門は、今日は米1粒、翌日には倍の2粒、その翌日には更に倍の4粒と、日ごとに倍の量の米を100日間もらう事を希望した。米粒なら大した事はないと思った秀吉は簡単に承諾したが、日ごとに倍ずつ増やして行くと100日後には膨大な量になる事に途中で気づき、他の褒美に変えてもらった。

### 【問題1】100日後の米粒の数
　100日後には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```python
day = 100 # 褒美をもらえる期間(日)
list_n_grains = [] # n日目にもらう米粒のリスト
list_total_grains = [] # 1日目からn日までにもらった米粒の累計のリスト

# 期間中にもらう米粒の量を計算
for i in range(1, day+1):
    list_n_grains += [2 ** (i-1)]
    list_total_grains += [sum(list_n_grains)]
    
print("{}日目にもらう米粒:{}".format(len(list_n_grains), list_n_grains[-1]))
print("{}日目までにもらった累計米粒:{}".format(len(list_total_grains), list_total_grains[-1]))

plt.title("Acumulated amount of grains of the day")
plt.xlabel("days")
plt.ylabel("amount of grains")
plt.plot(list_n_grains, color="b", label="amount of grains of the day")
plt.plot(list_total_grains, color="r", label="Acumulated amount of grains until the day")
plt.legend()
```

### 【問題3】何人が何日生活できる米か
　この米があれば何人が何日生活できるか計算する関数を作成してください。コースを受講する期間に貰える米粒で、このクラスにいる人数であれば、何日間生活できるかを出力してみましょう。関数の最低限の入力は 米粒の数(int) と 人数(int) 、出力は 生活できる日数(int) です。関数名やコメントなどは分かりやすく考えて作成してください。ここでは1日に必要な米粒の数は提示しません。自分なりに調べて、計算式に落とし込んでください。どう調べ、何を根拠にしたかもコメントを残しておいてください。

```python
def compute_sorori_shinzaemon(day=100):
    list_n_grains = [] # n日にもらう米粒のリスト
    list_total_grains = [] # 1日目からn日までにもらった累計の米粒の量のリスト

    # 期間中にもらう米粒の量を計算
    for i in range(1, day+1):
        list_n_grains += [2 ** (i-1)]
        list_total_grains += [sum(list_n_grains)]
    
    return int(list_total_grains[-1])

def calculate_survival_days(amount_grains, mens_persons, womens_persons):
    
    """
    農林水産省のウェブサイト参照　http://www.maff.go.jp/j/syokuiku/zissen_navi/balance/required.html
    ・一日に必要なエネルギー量（成人女性）：1400～2000kcal
    ・一日に必要なエネルギー量（成人男性）：2200～2400kcal
    """
    MENS_REQUIRED_KCAL = 2400 # 成人男性が1日に最低限必要なエネルギー量(kcal) 
    WOMENS_REQUIRED_KCAL = 2000 # 成人女性が1日に最低限必要なエネルギー量(kcal)
    
    """
    公益社団法人 米穀安定供給確保支援機構ウェブサイト参照　http://www.komenet.jp/_qa/chawanippai/index.html
    ・茶碗1杯のご飯(炊いた後)の重量(150g)
    ・茶碗1杯の精米(炊く前)の重量(65g)
    ・茶碗1杯の米粒の量(3250粒)
    ・茶碗1杯のご飯のエネルギー量(252kcal)
    """
    COOKED_RICE_WEIGHT= 150 # 茶碗1杯のご飯(炊いた後)の重量(g)
    MILLED_RICE_WEIGHT = 65 # 茶碗1杯の精米(炊く前)の重量(g)
    AMOUNT_GRAINS = 3250 # 茶碗1杯の米粒の量(粒)
    COOKED_RICE_KCAL = 240 # 茶碗1杯のご飯のエネルギー量(kcal)
    GRAIN_KCAL = AMOUNT_GRAINS / COOKED_RICE_KCAL # 米1粒当たりのエネルギー量(kcal)
    
    total_required_kcal = ((MENS_REQUIRED_KCAL * mens_persons) + (WOMENS_REQUIRED_KCAL * womens_persons)) # 合計必要エネルギー量(kcal)
    
    total_grains_kcal = amount_grains * GRAIN_KCAL # 米粒n個でのエネルギー量(kcal)
    survival_days = total_grains_kcal / total_required_kcal # 米粒n個でメンバーが生活できる日数
    
    return int(survival_days)


start_date = datetime.date(2019, 7, 1) # 受講開始日の日付
end_date = datetime.date(2019, 10, 31) # 受講終了日の日付
period_days = ((end_date - start_date).days + 1) # 開始日から終了日までの日数

mens_persons = 11 # 同期の男性人数
womens_persons = 2 # 同期の女性人数

compute_sorori_shinzaemon(period_days)

total_grains = compute_sorori_shinzaemon(period_days) # 受講期間でもらえる米粒の総数

survival_days = calculate_survival_days(total_grains, mens_persons, womens_persons) # 生活できる日数

print("もらった米粒でコース受講期間中に同期メンバーは {} 日間生活できる".format(survival_days))
```
